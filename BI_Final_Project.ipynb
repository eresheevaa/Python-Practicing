```python
import numpy as np
import pandas as pd

# Load in the dataframe
df = pd.read_csv("/content/Tennis.csv", index_col=0)
```

```python
# Looking at first 10 rows of the dataset
df.head(10)
```

```python
datatype = df.dtypes
print(datatype)
```

```python
# convert the 'Date' column to datetime format
df['Birthdate']= pd.to_datetime(df['Birthdate'])
```

```python
# Check the format of 'Date' column
df.dtypes
```

```python
#printout some basic information about the dataset using print() combined with .format() to have a nice printout.
print("There are {} observations and {} features in this dataset. \n".format(df.shape[0],df.shape[1]))

print("There are {} mentioned countries in this dataset such as {}... \n".format(len(df.Country.unique()),
                                                                                      ", ".join(df.Country.unique()[0:5])))
```

```python
# Groupby by country
country = df.groupby("Country")

# Summary statistic of all countries
country.describe().head()
```

```python
#this selects the top 5 highest average points among all 34 countries:
country.mean().sort_values(by="Points",ascending=False).head()
```

```python
#You can plot the number of players by country using the plot method of Pandas DataFrame and Matplotlib.

import matplotlib.pyplot as plt 

plt.figure(figsize=(15,10))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Players")
plt.show()
```

```python
#Let's now take a look at the plot of all 34 countries by its highest rated player, using the same plotting technique as above:
plt.figure(figsize=(15,10))
country.max().sort_values(by="Points",ascending=False)["Points"].plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Highest point")
plt.show()
```
